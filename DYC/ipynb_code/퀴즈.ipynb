```python
바탕질환= pd.read_csv('D:/Programming/Jupyter/02.data/데청캠/전처리데이터/바탕질환_최종.csv', encoding ='cp949')
진단명= pd.read_csv('D:/Programming/Jupyter/02.data/데청캠/진단명리스트.csv')
```

```python
항생제 = pd.read_csv('D:/Programming/Jupyter/02.data/데청캠/전처리데이터/항생제_투여.csv')
```

```python
바탕질환 = 바탕질환.drop(['주진단 여부'], axis=1)
바탕질환.head()
```

```python
항생제 = 항생제[['내원번호', '등록번호', '약품명','성분명']]
항생제.head()
```

```python
merged_df = pd.merge(바탕질환, 항생제, on=['내원번호', '등록번호'], how='inner')
merged_df = merged_df.drop(['내원번호','등록번호'], axis=1)
merged_df.head()
```

```python
# 진단명 열을 기준으로 두 데이터프레임을 합칩니다.
result_df = pd.merge(merged_df, 진단명, on='진단명')

result_df = result_df.drop(['진단명','카운트'], axis=1)
result_df.head()
```

```python
df = result_df
```

```python
# DataFrame을 엑셀 파일로 저장
file_path = r'D:\Programming\Jupyter\02.data\데청캠\퀴즈정리데이터.xlsx'
df.to_excel(file_path, index=False, engine='openpyxl')

print(f"DataFrame가 '{file_path}' 경로에 저장되었습니다.")
```

```python
import random

# 중복 데이터 제거
df = df.drop_duplicates()

# 랜덤 퀴즈 생성
def generate_random_quiz(df):
    # 데이터프레임에서 무작위로 하나의 행 선택
    random_row = df.sample(1)
    
    # 질문 및 정답 생성
    icd_10_code = random_row['ICD-10코드'].values[0]
    diagnosis_korean = random_row['진단명(한국어)'].values[0]
    symptoms = random_row['증상 키워드'].values[0].split(',')
    
    random.shuffle(symptoms)  # 증상 키워드를 섞습니다.
    
    question = f"다음 진단 코드에 해당하는 진단명(한국어)을 선택하세요: {icd_10_code}"
    options = [diagnosis_korean]
    
    # 랜덤하게 3개의 오답 추가
    while len(options) < 4:
        random_option = random.choice(df['진단명(한국어)'].unique())
        if random_option != diagnosis_korean and random_option not in options:
            options.append(random_option)
    
    random.shuffle(options)  # 옵션을 섞습니다.
    
    # 정답 옵션의 인덱스 찾기
    correct_index = options.index(diagnosis_korean)
    
    return question, options, correct_index

# 랜덤 퀴즈 생성
question, options, correct_index = generate_random_quiz(df)

# 퀴즈 출력
print(question)
for i, option in enumerate(options):
    print(f"{i + 1}. {option}")

# 정답 확인
user_answer = int(input("정답을 선택하세요 (숫자 입력): "))
if user_answer - 1 == correct_index:
    print("정답입니다!")
else:
    print("틀렸습니다. 정답은", correct_index + 1, "번입니다.")
```
